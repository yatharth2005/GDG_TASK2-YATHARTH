**CHECKING OUTLIERS USING ZSCORE**




```python
from scipy import stats
z= np.abs(stats.zscore(wine_df))
print(z)
```

```python
print(np.where(z>3))
```

```python
threshold =3

outliers = np.where(z > threshold)

outlier_count = (z > threshold).sum(axis=0)

# Display the number of outliers per feature
outlier_summary = dict(zip(wine_df.columns, outlier_count))
```

```python
outlier_summary
```

```python
new_df=wine_df[(z<3).all(axis=1)]
```
